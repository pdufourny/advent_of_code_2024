```python
list1 = []
list2 = []

file1 = open("file1.txt")
file2 = open("file2.txt")

list1 = file1.readlines()
list2 = file2.readlines()


file1.close
file2.close

liste_1 = [int(element.strip())   for element in list1]
liste_2 = [int(element.strip())   for element in list2]


```

```python
liste_1.sort()
liste_2.sort()
print(liste_1)
print(liste_2)
```

```python
res = 0
for i in range(len(liste_1)):
    distance = abs(liste_1[i] - liste_2[i])
    print(f"\t{i} : dist = {distance}")
    res +=  distance
print(f"\ndistance totale = {res}")
```

# Avec les données réelles

```python
lst_ = []
lst_1 = []
lst_2 = []
file = open("input_file.txt")
lst_ = file.readlines()
file.close

l = [ (int(el.split()[0]) , int(el.split()[1].strip()))  for el in lst_ ]
# l contains cleaned tuples corresponding to each line of the input file
lst_1, lst_2 = list(zip (*l))
# split the tuples into the two lists we need to process

# cast tuples to lists
lst_1 = list(lst_1)
lst_2 = list(lst_2)


lst_1.sort()
lst_2.sort()

# process the data
res = 0
for i in range(len(lst_1)):
    distance = abs(lst_1[i] - lst_2[i])
    print(f"\t{i} : dist = {distance}")
    res +=  distance
print(f"\ndistance totale = {res}")

```

```python
lst_ = []
lst_1 = []
lst_2 = []
file = open("input_file.txt")
lst_ = file.readlines()
file.close

l = [ (int(el.split()[0]) , int(el.split()[1].strip()))  for el in lst_ ]
# l contains cleaned tuples corresponding to each line of the input file
lst_1, lst_2 = list(zip (*l))
# split the tuples into the two lists we need to process

# cast tuples to lists
lst_1 = list(lst_1)
lst_2 = list(lst_2)

# process the data
similarity_score = 0

for i in lst_1:
    nb = lst_2.count(i)
    print("nbre de fois que ",i,"apparait dans la liste 2 est ",nb)
    similarity_score += nb * i



print(similarity_score)

```

```python
sum = 0
do_mul = True
for line in all_datas_split:
    match(line[0:4]):
        case "mul(":
            print("mul")
            if do_mul:
                pos = line.find(")")
                if pos>1:
                    args = line[4:pos].split(",")
                    if len(args) ==2 :
                        if args[0].isnumeric() and args[1].isnumeric():
                            if int(args[0])<1000 and int(args[1])<1000:
                                print(args[0],args[1])
                                sum += (int(args[0]) * int(args[1]))            
        case "do()":
            print("do")
            do_mul = True
        case "don'":
            print("don't")
            do_mul = False
        case _ :
            print("other")

print(sum) 
```

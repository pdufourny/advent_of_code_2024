```python
lst_ = []

file1 = open("input_file.txt")
lst_ = file1.readlines()
file1.close

liste_1 = [element.strip()   for element in lst_]

print(liste_1)
```

```python
liste_ok = []
for i in range(0,len(liste_1)):    
    st = liste_1[i].split()
    ec = int(st[0])
    for j in range(1,len(st)):
        print(f"{ec=}  {st[j]= } {j = }")
        if j<len(st)-1:
            if abs(ec-int(st[j]))>3 and abs(ec-int(st[j+1]))>3:
                print("ko")
                st=[]
                break
            else :
                ec = int(st[j])
        else :
            if abs(ec-int(st[j]))>3 :
                print("KO_1")
                st.pop(j)
                break

            

    print(f"new_st = {st=}")
    if len(st) > 0:
        #liste_1.pop(i)
        liste_ok.append(liste_1[i])

#print(liste_1)        



```
